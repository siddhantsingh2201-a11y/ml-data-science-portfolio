Learning Decision Tree

```python
# Adding the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
```

Loading the iris dataset.

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
```

```python
tree_clf = DecisionTreeClassifier(
    criterion= "gini",
    random_state= 0
)

tree_clf.fit(X_train, y_train)
```

```python
y_pred = tree_clf.predict(x_test)
```

```python
print(f"Test Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

Visualizing the results

```python
# The white box nature.
plt.figure(figsize=(12,8))
plot_tree(
    tree_clf,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled= True,
    rounded= True
         )
plt.title("Decision Tree Visualization")
plt.show()
```

```python
# Constrained (Pruned)
# Solution : max_depth = 3, min_samples_leaf = 5
tree_pruned = DecisionTreeClassifier(max_depth=3, min_samples_leaf=5,random_state = 42)
```

```python
tree_pruned.fit(X_train, y_train)
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, tree_pruned.predict(x_test)):.4f}")
```

```python
# Visualization
plt.figure(figsize=(12,8))
plot_tree(
    tree_pruned,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    rounded= True,
    filled= True
)
plt.title("Decision Tree (Overfitting Addressed)")
plt.show()
```
